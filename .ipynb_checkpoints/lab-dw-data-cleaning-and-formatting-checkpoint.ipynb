## Exercise 1: Cleaning Column Names

To ensure consistency and ease of use, standardize the column names of the dataframe. Start by taking a first look at the dataframe and identifying any column names that need to be modified. Use appropriate naming conventions and make sure that column names are descriptive and informative.

*Hint*:
- *Column names should be in lower case*
- *White spaces in column names should be replaced by `_`*
- *`st` could be replaced for `state`*

```python
import pandas as pd
import numpy as np
```

```python
customer_data=pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv')
display(customer_data.head())
```

```python
columns = []
for colname in customer_data.columns:
    columns.append(colname.lower())
    
customer_data.columns = columns

customer_data = customer_data.rename(columns={'st':'state',
                                'customer lifetime value':'customer_lifetime_value',
                                'monthly premium auto':'monthly_premium_auto',
                                 'number of open complaints':'number_of_open_complaints',
                                 'policy type':'policy_type',
                                 'vehicle class':'vehicle_class',
                                 'total claim amount':'total_claim_amount'})

display(customer_data.head())
```

## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they only contain valid data.

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

*Hint*:
- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*
- *State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
- *In education, "Bachelors" could be replaced by "Bachelor"*
- *In Customer Lifetime Value, delete the `%` character*
- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

```python
customer_data['gender'].replace('Femal', 'F', inplace=True)
customer_data['gender'].replace('female', 'F', inplace=True)
customer_data['gender'].replace('Male', 'M', inplace=True)
customer_data['state'].replace('Cali', 'California', inplace=True)
customer_data['state'].replace('AZ', 'Arizona', inplace=True)
customer_data['state'].replace('WA', 'Washington', inplace=True)
customer_data['education'].replace('Bachelors', 'Bachelor', inplace=True)
customer_data['vehicle_class'].replace('Sports Car', 'Luxury', inplace=True)
customer_data['vehicle_class'].replace('Luxury SUV', 'Luxury', inplace=True)
customer_data['vehicle_class'].replace('Luxury Car', 'Luxury', inplace=True)
```

```python
customer_data['vehicle_class'].unique()
```

## Exercise 3: Formatting data types

The data types of many columns in the dataset appear to be incorrect. This could impact the accuracy of our analysis. To ensure accurate analysis, we need to correct the data types of these columns. Please update the data types of the columns as appropriate.

It is important to note that this exercise does not involve handling null values (NaN or None).

*Hint*:
- *Customer lifetime value should be numeric*
- *Number of open complaints has an incorrect format. Look at the different values it takes with `unique()` and take the middle value. As an example, 1/5/00 should be 5. Number of open complaints is a string - remember you can use `split()` to deal with it and take the number you need. Finally, since it should be numeric, cast the column to be in its proper type.*

```python
# Check data type
print(customer_data.dtypes)
```

```python
# The replace() and astype() worked correct and the copy command was deleted in order to apply changes directly in the customer_data dataframe

customer_data['customer_lifetime_value'] = customer_data['customer_lifetime_value'].replace('%', '', regex=True).astype('float64')

# Display all data types with customer_lifetime_value modified to float64

display(customer_data.dtypes)
```

```python
# Display customer data dataframe with customer_lifetime_value cells without "%"

display(customer_data)
```

```python
# Change the format of open complaints

customer_data['number_of_open_complaints'].unique()
```

```python
# Split column "number_of_open_complaints" into 3 columns to remove the separate the number in the middle

customer_data[['first', 'second', 'third']] = customer_data['number_of_open_complaints'].str.split('/', expand=True)
display(customer_data)
```

```python
# Drop columns number_of_open_complaints, first, third

customer_data.drop(['number_of_open_complaints', 'first', 'third'], axis=1, inplace=True)
display(customer_data)
```

```python
# Rename column 'second' to 'number_of_open_complaints'

customer_data.rename(columns={"second": "number_of_open_complaints"}, inplace=True)
display(customer_data)
```

## Exercise 4: Dealing with Null values

Identify any columns with null or missing values. Identify how many null values each column has. You can use the `isnull()` function in pandas to find columns with null values.

Decide on a strategy for handling the null values. There are several options, including:

- Drop the rows or columns with null values
- Fill the null values with a specific value (such as the column mean or median for numerical variables, and mode for categorical variables)
- Fill the null values with the previous or next value in the column
- Fill the null values based on a more complex algorithm or model (note: we haven't covered this yet)

Implement your chosen strategy to handle the null values. You can use the `fillna()` function in pandas to fill null values or `dropna()` function to drop null values.

Verify that your strategy has successfully handled the null values. You can use the `isnull()` function again to check if there are still null values in the dataset.

Remember to document your process and explain your reasoning for choosing a particular strategy for handling null values.

After formatting data types, as a last step, convert all the numeric variables to integers using `applymap()`.

```python
# Review the customer_data dataframe

display(customer_data)
```

```python
# Drop all rows that only have missing values

customer_data = customer_data.dropna(how='all')
display(customer_data)
```

```python
# Check what's the amount of missing values in each column

customer_data.isna().sum()
```

```python
# Drop 3 rows with the customer_lifetime_value column as NaN. We will drop them as they are less than 0,003% of the rows.

rows_to_drop = customer_data[customer_data['customer_lifetime_value'].isna()==True].index
customer_data = customer_data.drop(rows_to_drop, axis=0)
customer_data.isna().sum()

# One of the rows that was dropped also had a NaN value on the column "gender"
```

```python
# 10% of the dataframe rows have NaN values in the column 'gender'. Therefore, I will fill the cells with the letter "U" from "Undefined" (instead of dropping them)

customer_data['gender'] = customer_data['gender'].fillna('U')

# Check the unique values for the column 'gender'

customer_data['gender'].unique()
```

```python
# Check variable types and convert all numeric variables to integers

customer_data.dtypes
```

```python

numeric_cols = customer_data.select_dtypes('number')

lst_cols = numeric_cols.columns

customer_data[lst_cols] = customer_data[lst_cols].applymap(np.int64)
customer_data.dtypes
```
